```csharp
%%sql
--Create TOMPo tables
use tompo;

CREATE TABLE IF NOT EXISTS tompo_datasources
(    
    ModelName STRING
   , ID BIGINT
   , ModelID BIGINT
   , Name STRING 
   , Description STRING
   , Type BIGINT
   , ConnectionString STRING
   , ImpersonationMode BIGINT
   , Account STRING
   , ModifiedTime TIMESTAMP
)
USING DELTA
LOCATION "/data/tompo/tompo_datasources";

CREATE TABLE IF NOT EXISTS tompo_tables
(    
     ModelName STRING
   , ID BIGINT
   , ModelID BIGINT
   , Name STRING 
   , DataCategory STRING
   , Description STRING
   , IsHidden BOOLEAN
)
USING DELTA
LOCATION "/data/tompo/tompo_tables";


CREATE TABLE IF NOT EXISTS tompo_tablepartitions
(    
     ModelName STRING
   , ID BIGINT
   , TableID BIGINT
   , Name STRING 
   , Description STRING
   , DataSourceID INT
   , QueryDefinition STRING
   , Type BIGINT
   , Mode BIGINT
   , ModifiedTime TIMESTAMP
)
USING DELTA
LOCATION "/data/tompo/tompo_tablepartitions";

CREATE TABLE IF NOT EXISTS tompo_totables
(    
     ModelName STRING 
   , ID BIGINT
   , ModelID BIGINT
   , Name STRING 
   , DataCategory STRING
   , Description STRING
   , IsHidden BOOLEAN
)
USING DELTA
LOCATION "/data/tompo/tompo_totables";


CREATE TABLE IF NOT EXISTS tompo_columns
(    
     ModelName STRING
   , ID BIGINT
   , TableID BIGINT
   , Name STRING    
   , ExplicitDataType STRING
   , DataCategory STRING
   , Description BOOLEAN
   , isHidden BOOLEAN
   , isUnique BOOLEAN
   , isKey BOOLEAN
   , SummarizeBy BIGINT
   , ColumnStorageID BIGINT
   , Type BIGINT
   , SourceColumn STRING
   , Expression STRING
   , FormatString STRING
   , SortByColumnID BIGINT
   , AttributeHierarchyID BIGINT
   , ModifiedTime TIMESTAMP
   , StructuredModifiedTime TIMESTAMP
   , DisplayFolder STRING
)
USING DELTA
LOCATION "/data/tompo/tompo_columns";


CREATE TABLE IF NOT EXISTS tompo_measures
(    
     ModelName STRING
   , ID BIGINT
   , TableID BIGINT
   , Name STRING
   , Description STRING    
   , DataType BIGINT
   , Expression STRING
   , FormatString STRING
   , IsHidden BOOLEAN
   , ModifiedTime TIMESTAMP
   , StructuredModifiedTime TIMESTAMP
   , KPID BIGINT
   , IsSimpleMeasure BOOLEAN
   , DisplayFolder STRING
)
USING DELTA
LOCATION "/data/tompo/tompo_measures";

CREATE TABLE IF NOT EXISTS tompo_calcdependency
(    
     ModelName STRING
   , Database_Name STRING
   , Object_Type STRING
   , Table STRING
   , Object STRING    
   , Expression STRING
   , ReferenceObjectType STRING
   , ReferencedTable STRING
   , ReferencedObject STRING
   , ReferencedExpression STRING
   , Query STRING
)
USING DELTA
LOCATION "/data/tompo/tompo_calcdependency";


CREATE TABLE IF NOT EXISTS tompo_relationships
(    
     ModelName STRING
   , ID BIGINT
   , ModelID BIGINT
   , IsActive BIGINT
   , Type BIGINT    
   , CrossFilteringBehavior STRING
   , FromTableID STRING
   , FromColumnID STRING
   , FromCardinality STRING
   , ToTableID STRING
   , ToColumnID STRING
   , ToCardinality BIGINT
   , ModifiedTime TIMESTAMP
)
USING DELTA
LOCATION "/data/tompo/tompo_relationships";


CREATE TABLE IF NOT EXISTS tompo_rolememberships
(    
     ModelName STRING
   , ID BIGINT
   , RoleID BIGINT
   , MemberName STRING    
   , IdentityProvider STRING
   , ModifiedTime TIMESTAMP
)
USING DELTA
LOCATION "/data/tompo/tompo_rolememberships";


CREATE TABLE IF NOT EXISTS tompo_roles
(    
     ModelName STRING
   , ID BIGINT
   , Name STRING
   , Description STRING    
   , ModelPermission INT
   , ModifiedTime TIMESTAMP
)
USING DELTA
LOCATION "/data/tompo/tompo_roles";

CREATE TABLE IF NOT EXISTS tompo_dbschema_catalogs
(
    ModelName STRING
  , Description STRING
  , ROLES STRING
  , DATE_MODIFIED TIMESTAMP
  , COMPATIBILITY_LEVEL BIGINT
  , TYPE STRING
)
USING DELTA
LOCATION "/data/tompo/tompo_dbschema_catalogs";


CREATE TABLE IF NOT EXISTS tompo_reportimpact
(    
     ModelName STRING
   , TableName STRING
   , ObjectName STRING
   , ReportName STRING
   , PageName STRING
   , VisualType STRING
   , ObjectType STRING
)
USING DELTA
LOCATION "/data/tompo/tompo_reportimpact";
```

```csharp
#r "nuget: Microsoft.AnalysisServices.AdomdClient.NetCore.retail.amd64, 19.51.0"

using System.Data;
using Microsoft.Spark.Sql;
using Microsoft.Spark.Sql.Types;
using T=Microsoft.Spark.Sql.Types;
using Microsoft.AnalysisServices.AdomdClient;

DataFrame RunXmlaQuery(string constr, string query, int topRows=0)
{
    using (var con = new AdomdConnection(constr))
    {
        con.Open();
        var cmd = con.CreateCommand();
        cmd.CommandText = query;
        using (var rdr = cmd.ExecuteReader())
        {
            return GetDataFrame(rdr,topRows);
        }

    }
}

DataFrame GetDataFrame( IDataReader rdr, int topRows = 0)
{
    var fields= GetFields(rdr).ToList();
    var type = new StructType(fields);

    //Console.WriteLine(type.SerializeToJson());

    return spark.CreateDataFrame(GetRows(rdr, topRows), type);
}

IEnumerable<GenericRow> GetRows(IDataReader rdr, int topRows = 0)
{    
    int rows = 0;
    while (rdr.Read())
    {
        rows++;
        var values = new object[rdr.FieldCount];
        rdr.GetValues(values);

        for (int i=0;i<rdr.FieldCount;i++)
        {
            var obj = values[i];
            if (obj is UInt64 u)
            {
                 values[i] = Convert.ToInt64(u);
            }
            else if (obj is DateTime dt)
            {
           
               values[i] = dt.ToString();
               
            }
            else if (obj is Date d)
            {
               
                values[i] = d.ToString();
               
            }            
        }
        
        yield return new GenericRow(values);

        if (topRows > 0 && rows >= topRows)
            break;

    }
}

DataType GetSparkType(Type t)
{
    if (t == typeof(string)) 
        return new T.StringType();
    if (t == typeof(int))
        return new T.IntegerType();
    if (t == typeof(long) || t == typeof(UInt64))
        return new T.LongType();
    if (t == typeof(float))
        return new T.FloatType();
    if (t == typeof(double))
        return new T.DoubleType();
    if (t == typeof(decimal))
        return new T.DecimalType();
    if (t == typeof(DateTime) || t == typeof(DateTimeOffset))
        //return new T.TimestampType(); not yet supported
        return new T.StringType();
    if (t == typeof(Date))
        //return new T.DateType();
        return new T.StringType();
    if (t == typeof(Guid))
        return new T.StringType();
    if (t == typeof(decimal))
        return new T.DecimalType();
    if (t == typeof(long))
        return new T.LongType();
    if (t == typeof(short))
        return new T.ShortType();
    if (t == typeof(bool))
        return new T.BooleanType();
    if (t == typeof(byte))
        return new T.ByteType();


    throw new InvalidOperationException($"Unsupported Type for DataFrame conversion: {t.Name}");
}

IEnumerable<StructField> GetFields(IDataReader rdr)
{
    for (int i = 0; i < rdr.FieldCount; i++)
    {
        var type = GetSparkType(rdr.GetFieldType(i));   
        var name = rdr.GetName(i);

        //Console.WriteLine($"{name} {type}");

        yield return new StructField(name, type, isNullable:true);
    }

}
```

```csharp
%%sql
USE tompo;
INSERT INTO tompo.tompo_reportimpact
select distinct a.ModelName, b.Name as TableName, c.Object as MeasureName, po.ReportName,po.PageName, po.VisualType, 'Measure'
from tompo.tompo_dbschema_catalogs a
JOIN tompo.tompo_tables b
on a.ModelName= b.ModelName
JOIN tompo.tompo_calcdependency c
on concat(b.ModelName,b.Name) = concat(c.ModelName,c.ReferencedTable)
JOIN vwParserOutput po
on concat(c.ModelName, c.Object) = concat(po.ModelName,po.Column)

UNION ALL

select DISTINCT a.ModelName, b.Name as TableName, c.Name as ColumnName,po.ReportName,po.PageName, po.VisualType, 'Column'
FROM tompo.tompo_dbschema_catalogs a
JOIN tompo.tompo_tables b
ON a.ModelName= b.ModelName
JOIN tompo.tompo_columns c
ON concat(b.ModelName, b.ID) = concat(c.ModelName,c.TableID)
JOIN vwParserOutput po
on concat(c.ModelName, c.Name) = concat(po.ModelName,po.Column)
```
